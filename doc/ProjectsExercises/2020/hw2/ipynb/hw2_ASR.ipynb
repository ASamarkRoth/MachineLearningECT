Ridge regression: 

$$\mathbf{\beta^{\mathrm{Ridge}}} = (\mathbf{X}^{T} \mathbf{X}+\lambda\mathbf{I})^{-1}\mathbf{X}^{T}\mathbf{y}$$

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.linear_model

x = np.random.rand(100,1)
y = 5*x*x+0.1*np.random.randn(100, 1)
df = pd.DataFrame(data=np.hstack([x, y]), columns=["x", "y"])
df.sort_values("x", inplace=True)
#df.plot("x", "y")
```

Design matrix

```python
p = 3 #second degree polynomial
X = np.empty((df.shape[0], p))
X[:,0] = 1
X[:,1] = df["x"]
X[:,2] = np.power(df["x"], 2)
DesignMatrix = pd.DataFrame(X)
DesignMatrix.columns = ['beta0', 'beta1', 'beta2']
DesignMatrix.set_index(df["x"], inplace=True)
#display(DesignMatrix)
```

## 1 Performing OLS, Ridge and Lasso fits for a variation of $\lambda$

Choosing $\lambda$ to investigate: 

> According to Friedman, Hastie & Tibshirani (2010) 'strategy is to select a minimum value lambda_min = epsilon * lambda_max, and construct a sequence of K values of lambda decreasing from lambda_max to lambda_min on the log scale. Typical values are epsilon = 0.001 and K = 100.'

```python
#from lecture slides
nlambdas = 500
lambdas = np.logspace(-3, 5, nlambdas)
Identity = np.identity(p)

ols_reg = sklearn.linear_model.LinearRegression().fit(X, df["y"])
sigma2 = (1/(X.shape[0]-p-1))*np.sum(np.square(df["y"]-ols_reg.predict(X)))
ols_cov = np.linalg.inv(X.T @ X)*sigma2

print("OLS:\nCoeff:", ols_reg.coef_, "\nVariance of beta:", np.diag(ols_cov))

beta_ridge = np.empty((nlambdas, 3))
beta_lasso = np.empty((nlambdas, 3))

beta_var_ridge = np.empty((nlambdas, 3))
beta_var_lasso = np.empty((nlambdas, 3))

R2_ridge = np.empty(nlambdas)
MSE_ridge = np.empty(nlambdas)
R2_lasso = np.empty(nlambdas)
MSE_lasso = np.empty(nlambdas)

for i, lmb in enumerate(lambdas):
    ridge = sklearn.linear_model.Ridge(alpha=lmb).fit(X, df["y"])
    lasso = sklearn.linear_model.Lasso(alpha=lmb).fit(X, df["y"])

    R2_ridge[i] = ridge.score(X, df["y"])
    R2_lasso[i] = lasso.score(X, df["y"])
    
    MSE_ridge[i] = sklearn.metrics.mean_squared_error(df["y"], ridge.predict(X))
    MSE_lasso[i] = sklearn.metrics.mean_squared_error(df["y"], lasso.predict(X))
    
    sigma2 = (1/(X.shape[0]-p-1))*np.sum(np.square(df["y"]-ridge.predict(X)))
    ridge_cov = np.linalg.inv(X.T @ X + lmb*Identity)*sigma2
    
    beta_ridge[i, :] = ridge.coef_
    beta_lasso[i, :] = lasso.coef_
    
    #print(ridge.coef_)
    beta_var_ridge[i, :] = np.diagonal(ridge_cov)
    beta_var_lasso[i, :] = np.diagonal(lasso_cov)
```

```python
plt.figure()
plt.plot(np.log10(lambdas), beta_ridge[:, 0], label = 'beta0 Ridge')
plt.plot(np.log10(lambdas), beta_ridge[:, 1], label = 'beta1 Ridge')
plt.plot(np.log10(lambdas), beta_ridge[:, 2], label = 'beta2 Ridge')

plt.plot(np.log10(lambdas), beta_lasso[:, 0], label = 'beta0 Lasso')
plt.plot(np.log10(lambdas), beta_lasso[:, 1], label = 'beta1 Lasso')
plt.plot(np.log10(lambdas), beta_lasso[:, 2], label = 'beta2 Lasso')

plt.xlabel('log10(lambda)')
plt.ylabel('Estimates')
plt.legend()
plt.show()
```

<p>
A much used approach before starting to train the data is  to preprocess our
data. Normally the data may need a rescaling and/or may be sensitive
to extreme values. Scaling the data renders our inputs much more
suitable for the algorithms we want to employ.

<p>
<b>Scikit-Learn</b> has several functions which allow us to rescale the
data, normally resulting in much better results in terms of various
accuracy scores.  The <b>StandardScaler</b> function in <b>Scikit-Learn</b>
ensures that for each feature/predictor we study the mean value is
zero and the variance is one (every column in the design/feature
matrix).  This scaling has the drawback that it does not ensure that
we have a particular maximum or minimum in our data set. Another
function included in <b>Scikit-Learn</b> is the <b>MinMaxScaler</b> which
ensures that all features are exactly between \( 0 \) and \( 1 \). The

<p>
The <b>Normalizer</b> scales each data
point such that the feature vector has a euclidean length of one. In other words, it
projects a data point on the circle (or sphere in the case of higher dimensions) with a
radius of 1. This means every data point is scaled by a different number (by the
inverse of it&#8217;s length).
This normalization is often used when only the direction (or angle) of the data matters,
not the length of the feature vector.

<p>
The <b>RobustScaler</b> works similarly to the StandardScaler in that it
ensures statistical properties for each feature that guarantee that
they are on the same scale. However, the RobustScaler uses the median
and quartiles, instead of mean and variance. This makes the
RobustScaler ignore data points that are very different from the rest
(like measurement errors). These odd data points are also called
outliers, and might often lead to trouble for other scaling
techniques.

### OLS as function of added complexity with bootstrapping

```python
n = 100
# Make data set.
x = np.linspace(-3, 3, n).reshape(-1, 1)
y = np.exp(-x**2) + 1.5 * np.exp(-(x-2)**2)+ np.random.normal(0, 0.1, x.shape)
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size=0.2)

n_bootstraps = 100
maxdegree = 15
    
error = np.zeros(maxdegree)
bias = np.zeros(maxdegree)
variance = np.zeros(maxdegree)
polydegree = np.zeros(maxdegree)

for degree in range(maxdegree):
    model = sklearn.pipeline.make_pipeline(sklearn.preprocessing.PolynomialFeatures(degree=degree), sklearn.linear_model.LinearRegression(fit_intercept=True))
    y_pred = np.empty((y_test.shape[0], n_bootstraps))

    for i in range(n_bootstraps):
        x_, y_ = sklearn.utils.resample(x_train, y_train)
        y_pred[:, i] = model.fit(x_, y_).predict(x_test).ravel()

    print(y_test.shape, y_pred.shape)
    polydegree[degree] = degree
    error[degree] = np.mean( np.mean((y_test - y_pred)**2, axis=1, keepdims=True) )
    bias[degree] = np.mean( (y_test - np.mean(y_pred, axis=1, keepdims=True))**2 )
    variance[degree] = np.mean( np.var(y_pred, axis=1, keepdims=True) )
    print('Polynomial degree:', degree)
    print('Error:', error[degree])
    print('Bias^2:', bias[degree])
    print('Var:', variance[degree])
    print('{} >= {} + {} = {}'.format(error[degree], bias[degree], variance[degree], bias[degree]+variance[degree]))

plt.plot(polydegree, error, label='Error')
plt.plot(polydegree, bias, label='bias')
plt.plot(polydegree, variance, label='Variance')
plt.yscale('log')
plt.legend()
plt.show()
```

### Optimising hyperparameter for Ridge regression

```python
n = 100
# Make data set.
x = np.linspace(-3, 3, n).reshape(-1, 1)
y = np.exp(-x**2) + 1.5 * np.exp(-(x-2)**2)+ np.random.normal(0, 0.1, x.shape)
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size=0.2)

# Decide which values of lambda to use
nlambdas = 500
lambdas = np.logspace(-3, 5, nlambdas)

estimated_mse_sklearn = np.zeros(nlambdas)
degree = 20 # from OLS estimates
for i, lmb in enumerate(lambdas):
    model = sklearn.pipeline.make_pipeline(sklearn.preprocessing.PolynomialFeatures(degree=degree), sklearn.linear_model.Ridge(alpha=lmb, fit_intercept=True))
    clf_ridge = model.fit(x_train, y_train)
    yridge = clf_ridge.predict(x_test)
    estimated_mse_sklearn[i] = sklearn.metrics.mean_squared_error(y_test, yridge)
plt.figure()
plt.plot(np.log10(lambdas), estimated_mse_sklearn, label = 'Ridge MSE')
plt.xlabel('log10(lambda)')
plt.ylabel('MSE')
plt.legend()
plt.show()
```

### Ridge as function of added complexity with scaled data

```python
n = 100
# Make data set.
x = np.linspace(-3, 3, n).reshape(-1, 1)
y = np.exp(-x**2) + 1.5 * np.exp(-(x-2)**2)+ np.random.normal(0, 0.1, x.shape)
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size=0.2)

# choose type of scaler
scaler = sklearn.preprocessing.StandardScaler()
#scaler = sklearn.preprocessing.Normalizer()
scaler = sklearn.preprocessing.RobustScaler()

scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)

n_bootstraps = 100
maxdegree = 15
    
error = np.zeros(maxdegree)
bias = np.zeros(maxdegree)
variance = np.zeros(maxdegree)
polydegree = np.zeros(maxdegree)

lmb = 0.5 #from degree == 10 of above

for degree in range(maxdegree):
    model = sklearn.pipeline.make_pipeline(sklearn.preprocessing.PolynomialFeatures(degree=degree), sklearn.linear_model.Ridge(alpha=lmb, fit_intercept=False))
    y_pred = np.empty((y_test.shape[0], n_bootstraps))
    for i in range(n_bootstraps):
        x_, y_ = sklearn.utils.resample(x_train_scaled, y_train)
        y_pred[:, i] = model.fit(x_, y_).predict(x_test_scaled).ravel()

    polydegree[degree] = degree
    error[degree] = np.mean( np.mean((y_test - y_pred)**2, axis=1, keepdims=True) )
    bias[degree] = np.mean( (y_test - np.mean(y_pred, axis=1, keepdims=True))**2 )
    variance[degree] = np.mean( np.var(y_pred, axis=1, keepdims=True) )
    print('Polynomial degree:', degree)
    print('Error:', error[degree])
    print('Bias^2:', bias[degree])
    print('Var:', variance[degree])
    print('{} >= {} + {} = {}'.format(error[degree], bias[degree], variance[degree], bias[degree]+variance[degree]))

plt.plot(polydegree, error, label='Error')
plt.plot(polydegree, bias, label='bias')
plt.plot(polydegree, variance, label='Variance')
plt.yscale('log')
plt.legend()
plt.show()
```
